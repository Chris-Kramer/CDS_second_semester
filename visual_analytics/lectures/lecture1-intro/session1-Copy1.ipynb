## Image translation

```python
path_to_image = os.path.join("..", "data", "img", "trex.png")
image = cv2.imread(path_to_image)
jimshow(image, "Original")
```

Translation means moving the picture

__SHifting image down and right__

Translation matrix

Shift 25 pixels x-axis, 50 on y
M is short for matrix

```python
M = np.float64([[1, 0, 25],
                [0, 1, 50]])
```

```python
width = image.shape[1]
height = image.shape[0]
```

```python
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "Shifted down and right")
```

__Shifting up and left__

Shift 100 pixels left and 100 pixels up

```python
M = np.float64([[1, 0, -100],
                [0, 1, -100]])
```

```python
shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "Shifted up and left")
```

```python
M = np.float64([[1, 0, 0],
                [0, 1, -100]])

shifted = cv2.warpAffine(image, M, (width, height))
```

```python
jimshow(shifted, "Shifted down")
```

__Writing a function for images__

```python
def shift_img(image, x = 50, y = 75):
    # Define translation matrix
    M = np.float64([[1, 0, x],[0, 1, y]])
    #Save height and width in variables
    width = image.shape[1]
    height = image.shape[0]
    #Perform translation
    shifted = cv2.warpAffine(image, M, (width, height))
    return shifted
```

```python
new_image = shift_img(image, 0, 100)
jimshow(new_image)
```

```python
defaul_parameters = shift_img(image)
jimshow(defaul_parameters, "default parameters")
```
